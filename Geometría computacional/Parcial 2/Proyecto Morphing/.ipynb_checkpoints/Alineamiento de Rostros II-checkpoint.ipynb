En esta libreta exploraremos el *alineamiento* de rostros diferentes. 

Iniciaremos leyendo las im√°genes sobre las que trabajaremos y cargando algunas funciones necesarias para el proceso. 

```python
import numpy as np
import matplotlib.pyplot as plt
import dlib
import cv2
```

```python
def rect_to_bb(rect):
    x = rect.left()
    y = rect.top()    
    w = rect.right() - x
    h = rect.bottom() - y
    
    return (x, y, w, h)

def shape_to_np(shape, dtype="int"):
    coords = np.zeros((68, 2), dtype=dtype)
 
    for i in range(0, 68):
        coords[i] = (shape.part(i).x, shape.part(i).y)
    
    return coords

def rect2homog(pts):
    n = pts.shape[0]
    o = np.ones(n)
    return np.column_stack((pts,o))

def aplicaHomografia(pts, H):
    pts_h = rect2homog(pts)
    pts_t = np.dot(H,np.transpose(pts_h))
    pts_t_trans = np.transpose(pts_t)
    return normalizaPuntos(pts_t_trans)

def normalizaPuntos(pts_h):
    return np.column_stack((np.divide(pts_h[:,0],pts_h[:,2]),np.divide(pts_h[:,1],pts_h[:,2])))
```

```python
im1 = cv2.imread("mona.jpg",0)
im2 = cv2.imread("marilyn.jpg",0)

plt.subplot(121)
plt.imshow(im1, cmap="gray")
plt.axis("off")
plt.subplot(122)
plt.imshow(im2, cmap="gray")
plt.axis("off")
plt.show()
```

Dado que las im√°genes no tienen las mismas dimensones, reescalaremos una de ellas.

```python
im2 = cv2.resize(im2,(im1.shape[1],im1.shape[0]))
```

```python
plt.subplot(121)
plt.imshow(im1, cmap="gray")
plt.axis("off")
plt.subplot(122)
plt.imshow(im2, cmap="gray")
plt.axis("off")
plt.show()
```

Iniciamos repitiendo el proceso de la libreta anterior e identificamos marcadores en ambos rostros.

```python
#construimos los detectores de rostros y de marcadores
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")
```

```python
rects1 = detector(im1) 
shape1 = predictor(im1, rects1[0])
shape1 = shape_to_np(shape1)
len(shape1)
```

```python
rects2 = detector(im2) 
shape2 = predictor(im2, rects2[0])
shape2 = shape_to_np(shape2)
len(shape2)
```

```python
#mostramos los marcadores
im_copy1 = im1.copy()
im_copy2 = im2.copy()

for (x, y) in shape1:
    cv2.circle(im_copy1, (x, y), 2, (0, 0, 255), -1)
    
for (x, y) in shape2:
    cv2.circle(im_copy2, (x, y), 2, (0, 0, 255), -1)
    

plt.subplot(121)
plt.imshow(im_copy1,cmap="gray")
plt.axis("off")

plt.subplot(122)
plt.imshow(im_copy2,cmap="gray")
plt.axis("off")

plt.show()
```

Una vez que identificamos puntos correspondientes, recuperamos la homograf√≠a y alineamos las im√°genes.

```python
pts1 = 1.0*shape1 # multiplicamos por 1.0 para forzar a que los datos tengan el tipo adecuado
pts2 = 1.0*shape2

h, v = cv2.findHomography(pts1,pts2)

imt = cv2.warpPerspective(im1,h,(im1.shape[1],im1.shape[0]))

plt.subplot(121)
plt.imshow(imt,cmap="gray")
plt.axis("off")
plt.subplot(122)
plt.imshow(im2,cmap="gray")
plt.axis("off")
plt.show()
```

Podemos realizar un alineamiento m√°s detallado si seccionamos ambos rostros en segmentos correspondientes y alineamos dichos segmentos. Para este fin, la triangulaci√≥n de Delaunay puede ser muy √∫til, pero antes agregaremos algunos puntos adicionales a los marcadores que hemos detectado.

```python
#agregamos puntos adicionales ubicados en el marco de ambas im√°genes
shape1 = np.append(shape1,[[im1.shape[1]-1,im1.shape[0]/2],[im1.shape[1]/2,im1.shape[0]-1],[im1.shape[1]/2,0],[0,im1.shape[0]/2],[0,0],[im1.shape[1]-1,0],[0,im1.shape[0]-1],[im1.shape[1]-1,im1.shape[0]-1]], axis=0)
len(shape1)
```

```python
shape2 = np.append(shape2,[[im2.shape[1]-1,im2.shape[0]/2],[im2.shape[1]/2,im2.shape[0]-1],[im2.shape[1]/2,0],[0,im2.shape[0]/2],[0,0],[im2.shape[1]-1,0],[0,im2.shape[0]-1],[im2.shape[1]-1,im2.shape[0]-1]], axis=0)
len(shape2)
```

# Triangulaci√≥n de Delaunay

Aplicamos la triangulaci√≥n de Delaunay sobre la imagen que usaremos como referencia

```python
from scipy.spatial import Delaunay
```

```python
tri = Delaunay(shape1)

cp_1 = im1.copy()

for t in tri.simplices:    
    pt0 = (shape1[t[0]][0],shape1[t[0]][1])
    pt1 = (shape1[t[1]][0],shape1[t[1]][1])
    pt2 = (shape1[t[2]][0],shape1[t[2]][1])
    
    cv2.line(cp_1,pt0,pt1,(255,255,255),1)
    cv2.line(cp_1,pt0,pt2,(255,255,255),1)
    cv2.line(cp_1,pt1,pt2,(255,255,255),1)
    
plt.imshow(cp_1,cmap="gray")
plt.axis("off")
plt.show()     
```

**Aplicamos** la triangulaci√≥n obtenida en la segunda imagen [observa que no construimos una nueva] demanera que los tri√°ngulos en esta segunda imagen corresponden a los de la primera.

```python
cp_2 = im2.copy()

for t in tri.simplices:    
    pt0 = (shape2[t[0]][0],shape2[t[0]][1])
    pt1 = (shape2[t[1]][0],shape2[t[1]][1])
    pt2 = (shape2[t[2]][0],shape2[t[2]][1])
    
    cv2.line(cp_2,pt0,pt1,(255,255,255),1)
    cv2.line(cp_2,pt0,pt2,(255,255,255),1)
    cv2.line(cp_2,pt1,pt2,(255,255,255),1)
    
plt.imshow(cp_2,cmap="gray")
plt.axis("off")
plt.show()      
```

**Observaci√≥n** Con tres correspondencias solamente podemos determinar una transformaci√≥n af√≠n.

```python
#dibuja el tri√°ngulo seleccionado sobre la segunda imagen 
cp_2 = im2.copy()

pt0 = (shape2[t[0]][0],shape2[t[0]][1])
pt1 = (shape2[t[1]][0],shape2[t[1]][1])
pt2 = (shape2[t[2]][0],shape2[t[2]][1])

pts2 = [pt0, pt1, pt2]

cv2.line(cp_2,pt0,pt1,(255,255,255),1)
cv2.line(cp_2,pt0,pt2,(255,255,255),1)
cv2.line(cp_2,pt1,pt2,(255,255,255),1)
```

```python
#el orden (pts2, pts1) es importante buscamos 
#determinar la transformaci√≥n que alinee la
#imagen 2 con la 1
taf = cv2.getAffineTransform(pts2,pts1)
imt = cv2.warpAffine(im2,taf,(im2.shape[1],im2.shape[0]))
```

# Transformaci√≥n af√≠n a pedazos

Para alinear los rostros debemos transformar cada tri√°ngulo de forma independiente. 

```python
#env√≠a los √≠ndices del tri√°ngulo, 
#los puntos del tri√°ngulo en I1, 
#la imagen a transformar
def triangulo_transf(t, shape1, shape2, im):
    
    pt0_1 = (shape1[t[0]][0],shape1[t[0]][1])
    pt1_1 = (shape1[t[1]][0],shape1[t[1]][1])
    pt2_1 = (shape1[t[2]][0],shape1[t[2]][1])
    pts1 = np.array([pt0_1, pt1_1, pt2_1])
    
    pt0_2 = (shape2[t[0]][0],shape2[t[0]][1])
    pt1_2 = (shape2[t[1]][0],shape2[t[1]][1])
    pt2_2 = (shape2[t[2]][0],shape2[t[2]][1])
    pts2 = np.array([pt0_2, pt1_2, pt2_2])
    
    pts1 = np.float32(pts1)
    pts2 = np.float32(pts2)
    
    #determina la transformaci√≥n af√≠n
    taf = cv2.getAffineTransform(pts2,pts1)        
    
    #determina la homograf√≠a
    H = np.append(taf,[[0,0,1]], axis=0)
       
    #transforma los puntos
    pts_t = aplicaHomografia(pts2,H)
    
    (m,n) = im.shape
    
    alpha = np.zeros((m,n))
    
    points = [[int(pts_t[0][1]-0.5),int(pts_t[1][1]-0.5), int(pts_t[2][1]-0.5)],[int(pts_t[0][0]-0.5), int(pts_t[1][0]-0.5), int(pts_t[2][0]-0.5)], [1,1,1]]
    
    for i in range(min(points[0]),max(points[0])):
        for j in range(min(points[1]),max(points[1])):            
            x = np.linalg.solve(points,[i,j,1])
            if min(x) > 0: #all coefficients positive
                alpha[i,j] = 1
    im_t = cv2.warpAffine(im,taf,(n,m))
    alpha[alpha>0] = im_t[alpha>0]
    return alpha
```

```python
#selecciono un tri√°ngulo
n = 10
t = tri.simplices[n]

#transforma la imagen alineando el tri√°ngulo y lo extrae
triang_tr = triangulo_transf(t, shape1, shape2, im2)

#muestra la imagen
plt.gray()
plt.subplot(121)
plt.axis("off")
plt.imshow(cp_1)

plt.subplot(122)
plt.imshow(triang_tr,cmap="gray")
plt.axis("off")
plt.show()
```

Transformamos cada tri√°ngulo

```python
im_fin = np.zeros((im1.shape))
for t in tri.simplices:    
    #transforma la imagen alineando el tri√°ngulo y lo extrae
    triang_tr = triangulo_transf(t, shape1, shape2, im2)
    im_fin = im_fin + triang_tr

#muestra la imagen
plt.gray()
plt.subplot(121)
plt.axis("off")
plt.imshow(im1)

plt.subplot(122)
plt.imshow(im_fin,cmap="gray")
plt.axis("off")
plt.show()
```
