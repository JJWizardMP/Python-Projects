```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

# Alineamiento de imágenes

Supón que tenemos dos imágenes de la misma escena pero tomadas en diferentes instantes de tiempo y, posiblemente, mediante instrumentos diferentes. Por ejemplo:

* Imágenes médicas obtenidas de la misma sección anatómica de un mismo paciente en diferentes momentos.
* Imágenes satelitales de [aproximadamente] una misma región.
* Imágenes consecutivas de una cámara montada sobre un robot móvil.

Suponemos que las imágenes son tales que **existe una homografía** que, aplicada a una de las imágenes puede transformarla para alinear los puntos correspondientes en ambas imágenes.  

Cargamos las imágenes a trabajar:

```python
def ShowImages(im1, im2):
    plt.subplot(1,2,1)
    plt.imshow(im1,cmap="gray")
    plt.axis("off")
    plt.subplot(1,2,2)
    plt.imshow(im2,cmap="gray")
    plt.axis("off")
    plt.show()

def ShowPoints(im, hull, ps):
    imc = im.copy()
    fig = plt.figure()
    ax = fig.add_subplot(111)
    for i,p in enumerate(hull):
        if i in ps:
            cv2.circle(imc,(p[0][0],p[0][1]),10,3,-1)
            if i%1==0:
                ax.annotate(str(i),xy=((p[0][0]+10,p[0][1]+10)))
    plt.imshow(imc,cmap="gray")
    plt.show()
    
def ConvexHull(im1, im2):
    ret, im_th1 = cv2.threshold(im1, 135, 255, 0)
    ret, im_th2 = cv2.threshold(im2, 135, 255, 0)

    cont1, jerarq = cv2.findContours(im_th1.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    cont2, jerarq = cv2.findContours(im_th2.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

    hull1 = cv2.convexHull(cont1[0])
    hull2 = cv2.convexHull(cont2[0])
    return (hull1, hull2)

def ShowConvexHull(im1, im2):
    hull1, hull2 = ConvexHull(im1, im2)
    im_h1 = im1.copy()
    im_h2 = im2.copy()

    cv2.drawContours(im_h1, [hull1], -1, (0, 0, 255), 3)
    cv2.drawContours(im_h2, [hull2], -1, (0, 0, 255), 3)

    plt.subplot(1,2,1)
    plt.imshow(im_h1,cmap="gray")
    plt.axis("off")
    plt.subplot(1,2,2)
    plt.imshow(im_h2,cmap="gray")
    plt.axis("off")
    plt.show()
    
# Función que alinea los puntos realizando el cálculo de las intesidades sobre las imágenes
def AlineaPuntos(im1, im2):
    hull1, hull2 = ConvexHull(im1, im2)
    pts = []
    w = 10
    r2 = w**2
    for hu1 in hull1:
        val = []
        for i2, hu2 in enumerate(hull2):
            (x1,y1) = (hu1[0][0], hu1[0][1])
            (x2,y2) = (hu2[0][0], hu2[0][1])

            ints1 = []
            ints2 = []
            for i in range(-w,w+1):
                for j in range(-w,w+1):
                    d2 = i**2+j**2
                    if d2 <= r2:
                        x1p = x1+i
                        y1p = y1+j
                        x2p = x2+i
                        y2p = y2+j
                        flag = [False, False]
                        # Verifica si el punto no se sale del borde de la primera imagen
                        f = np.logical_and([x1p > 0, y1p > 0], [x1p < im1.shape[1], y1p < im1.shape[0]])
                        flag[0] = np.logical_and(f[0], f[1])
                        # Verifica si el punto no se sale del borde de la segunda imagen
                        f = np.logical_and([x2p > 0, y2p > 0], [x2p < im2.shape[1], y2p < im2.shape[0]])
                        flag[1] = np.logical_and(f[0], f[1])
                        if np.logical_and(flag[0], flag[1]):
                            # Se verifica que la posición de la matriz sea diferente de 0
                            if im2[y2p, x2p] != 0:
                                # Guardar intesidades
                                ints1.append(float(im1[y1p,x1p]))
                                ints2.append(float(im2[y2p,x2p]))
            ints1.sort() 
            ints2.sort() 
            m = np.column_stack((ints1,ints2))
            # Promedio de la Diferencia de cuadrados
            nprom = np.mean((np.subtract(m[:,0],m[:,1]))**2)
            #Guardamos valores
            val.append((nprom, i2))
        #Ordenar valores
        val.sort()
        #Guardar indice
        pts.append(val[0][1])

    pts1 = []; pts2 = []
    # Obtenemos las coordenadas correspondientes a los puntos alineados
    pts1 = [[h[0][0], h[0][1]] for h in hull1 ]
    pts2 = [[hull2[i][0][0], hull2[i][0][1] ] for i in pts ]
    pts1 = np.array(pts1, dtype=float)
    pts2 = np.array(pts2, dtype=float)
    return (pts1, pts2)
```

```python
im1 = cv2.imread("1_0.png",0)
im2 = cv2.imread("I_2.png",0)

ShowImages(im1, im2)
```

Obtenemos las Envolventes Convexas de ambas imágenes.

```python
ShowConvexHull(im1, im2)
```

## Búsqueda de puntos correspondientes ( Inspección Visual)

* Buscamos correspondencias **a mano**

```python
# Índices de los puntos definidos con observación de los puntos de la Envolvente Convexa
pi1 = [ 1, 5, 8, 11, 15, 20, 26, 28, 30 ] 
pi2 = [ 0, 3, 5, 8, 11, 19, 25, 30, 34 ]
hull1, hull2 = ConvexHull(im1, im2)
# Listas de puntos para ambas imágenes
pts1 = []; pts2 = []
# Filtramos puntos correspondientes a los encontrados mediante el aspecto visual
pts1 = [[h[0][0], h[0][1]] for i,h in enumerate(hull1) if i in pi1 ]
pts2 = [[h[0][0], h[0][1]] for i,h in enumerate(hull2) if i in pi2 ]
# Convertimos la lista de puntos a un arreglo de tipo float
pts1 = np.array(pts1, dtype=float)
pts2 = np.array(pts2, dtype=float)
```

#### Puntos en la primera imagen.

```python
# [ 1,5,8,11,15,20,26,28,30 ] len = 9
ShowPoints(im1, hull1, pi1)
```

#### Puntos en la segunda imagen.

```python
# [ 0,3,5,8,11,19,25,30,34 ] len = 9
ShowPoints(im2, hull2, pi2)
```

#### Recuperación y aplicación de la homografía de la primera imagen en la segunda

```python
# Encontramos la homografía que alinea la primera imagen con la segunda
homo, c = cv2.findHomography(pts1, pts2)
print "Homografía:\n" + str(np.around(homo, decimals=2))
imt = cv2.warpPerspective(im1,homo,(im1.shape[1],im1.shape[0]))
ShowImages(imt, im2)
```

#### Recuperación y aplicación de la homografía de la segunda imagen en la primera

```python
# Encontramos la homografía que alinea la segunda imagen con la primera
homo, c = cv2.findHomography(pts2, pts1)
print "Homografía:\n" + str(np.around(homo, decimals=2))
imt = cv2.warpPerspective(im2,homo,(im2.shape[1],im2.shape[0]))
ShowImages(im1, imt)
```

## Búsqueda de puntos correspondientes ( Calculando Intensidades )

* Comparamos las **intensidades** de parejas de puntos

```python
ps1, ps2 = AlineaPuntos(im1, im2)
```

#### Recuperación y aplicación de la homografía de la primera imagen en la segunda

```python
# Encontramos la homografía que alinea la primera imagen con la segunda
homo, c = cv2.findHomography(ps1, ps2)
print "Homografía:\n" + str(np.around(homo, decimals=2))
imt = cv2.warpPerspective(im1,homo,(im1.shape[1],im1.shape[0]))
ShowImages(imt, im2)
```

#### Recuperación y aplicación de la homografía de la segunda imagen en la primera

```python
# Encontramos la homografía que alinea la segunda imagen con la primera
homo, c = cv2.findHomography(ps2, ps1)
print "Homografía:\n" + str(np.around(homo, decimals=2))
imt = cv2.warpPerspective(im2,homo,(im2.shape[1],im2.shape[0]))
ShowImages(im1, imt)
```

### Conclusión

<P  style="text-align:justify;">Identificando puntos críticos sobre la <i>Envolvente Convexa</i> de ambas imágenes de la mano derecha que coinciden respectivamente con los dedos de cada imagen podemos aplicar las funciones de <i>OpenCV</i> para recuperar y aplicar la homografía obtenida en ambos casos para poder alinear las imágenes, por lo que podemos visualizar sobre la alineación a primera vista se ve bien pero no es precisa ya que se puede notar que totalmente ambas manos no coinciden con total exactitud, lo puntos malos de realizar este método es que necesitamos conocer los puntos que se asemejan a la ubicación de la <i>Envolvente Convexa</i> sobre ambas imagenes y realizar dicha tarea puede ser un proceso tedioso dependiendo del método que apliquemos para conocer dichos puntos.</P>

<p style="text-align:justify;">Por otro lado realizando el cálculo de las intesidades de colores sobre los puntos en la segunda imagen podemos por medio del cálculo de la correlación obtener el promedio del punto en la segunda que se asemeje más al punto sobre la primera imagen, esto nos dara una alineación más precisa que el primer método usado debido a que ahora usamos todos los puntos que correspondes a las <i>Envolventes Convexas</i> de ambas imágenes, por lo que se tiene una mayor presición ya que los puntos se acomodan respecto a la tranformación geométrica aplicada, de igual forma obteniendo la alineación correcta de los puntos podemos aplicar las funciones de <i>OpenCV</i> para recuperar y aplicar la homografía que transforma la primera imagen a la segunda, y la segunda a la primera respectivamente. Aunque la aparencia de las imágenes difiere un poco en la transformación, la rotación es correcta ya que cuando se aplica la homografía recuperada dependiendo de la transformación ambas imágenes se encuentran alineadas correctamente.</p>
