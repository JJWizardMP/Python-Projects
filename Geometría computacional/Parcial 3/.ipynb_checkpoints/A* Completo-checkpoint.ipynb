```python
import matplotlib.pyplot as plt
import cv2
%matplotlib inline 
```

```python
def Plot(image, title):
    plt.title(title)
    plt.imshow(image,cmap="gray")
    plt.axis("off")
    plt.show()
```

```python
im = cv2.imread("index.jpeg", cv2.COLOR_BGR2RGB )
Plot(im, 'eltelacreistewexDDd (2016)')
```

```python
im = cv2.imread("ind.jpeg", cv2.COLOR_BGR2RGB )
Plot(im, 'meper Donas (2019)')
```
