# ¿Cómo construir la lista de bordes de los mapas sobrepuestos?

Tenemos la lista doblemente ligada de ambos mapas:

* $\mathcal{D}_1$
* $\mathcal{D}_2$

Iniciamos $\mathcal{D} = \mathcal{D}_1\cup \mathcal{D}_2$

Basándose en el algoritmo de barrido de plano construímos la lista doblemente ligada de la sobreposición de mapas: $\mathcal{D}$, las tablas de vértices, bordes y caras deben actualizarse (corregirse)

## Vértices 

1. Iniciar con la unión de los vértices de ambos mapas: $\mathcal{V}(\mathcal{D}) = \mathcal{V}(\mathcal{D_1})\cup \mathcal{V}(\mathcal{D_2})$. **Observar** Las coordenadas no cambian ni la referencia a algún semiborde adyacente.
2. Si el vértice es nuevo (es la intersección de bordes de ambos mapas) entonces se deben guardar:
    3. Las coordenadas
    4. Una referencia a algún semiborde adyacente

## Semibordes

1. Buscamos la intersección de los bordes de ambos mapas utilizando el **barrido de plano**.
2. Cuando se deba revisar un vértice, $\nu$:
    3. Si $\nu$ contiene solamente bordes de un mapa ($\mathcal{D}_1$ o $\mathcal{D}_2$):
        4. No se realizan modificaciones a $\mathcal{D}$
    5. En caso contrario:
        6. $\texttt{ManejarInterseccion}(\nu)$        

Algunas **consideraciones** para la función 

$\texttt{ManejarInterseccion}(\nu)$        

1. Para cada semiborde adyacente al vértice:
    2. Se generan dos nuevos semibordes con origen en el vértice.
3. Los semi-bordes que no tienen su origen en el vértice $\nu$ deben actualizar la referencia al **siguiente** semiborde.
4. Los nuevos semibordes tienen su origen en el vértice $\nu$ deben registrar la referencia a los semibordes **siguiente** y **previo** 
5. Para identificar los semibordes **siguiente** y **previo** considerar un ordenamiento de los semibordes adyacentes a un vértice en sentido horario (para el siguiente) o antihorario (para el previo).

## Caras

* El número de caras debe ser igual al de secuencias conexas de semibordes "externos" más una.

* Si la tabla de semibordes se actalizó correctamente con el algoritmo de barrido, entonces debemos identificar cuál es la cara incidente a cada semi-borde

* Podemos **identificar** si es una frontera **externa** o **interna** localizando el vértice más a la izquierda en la secuencia y verificando el ángulo entre los semibordes

<img src="inner_outer.png"/>


### Grafo de caras

Para decidir qué secuencias encierran una cara debemos construir un grafo

* Cada secuencia de semibordes (externa o interna) se representa por un nodo del grafo
* Existe una arista entre una pareja de nodos si uno de los nodos es una frontera interna y la secuencia del otro nodo tiene un semiborde inmediatamente a la izquierda del vértice en el extremo izquierdo de la secuencia interna

<img src="grafica.png"/>

### Pseudocódigo

$\texttt{SOBREPOSICION_MAPAS}(\mathcal{D}_1, \mathcal{D}_2)$        

**Entrada**: Dos listas de bordes doblemente conectados que representan a los mapas orginales
**Salida**: Lista de bordes doblemente conectados de la sobreposición de ambos mapas.

1. Unir las listas originales: $\mathcal{D} = \mathcal{D}_2\cup\mathcal{D}_2 $
2. Calcular las intersecciones entre los bordes de ambos mapas con el algoritmo de barrido de mapa. Considerar que, en cada evento, $\nu$, además actualizar el estado $\mathcal{T}$, y la lista de eventos $\mathcal{Q}$ es necesario realizar las siguientes acciones
    3. Actualizar $\mathcal{D}$ si el evento contiene bordes de ambos mapas. Considerar para esta actualización las ideas discutidas en  $\texttt{ManejarInterseccion}(\nu)$         
    4. Guardar el borde inmediatamente a la izquierda del evento.
3. A esta altura $\mathcal{D}$ debe contener la lista de sembibordes doblemente conectados excepto que la información de las caras no ha sido calculada aún.
4. Identificar cada secuencia de semibordes en $\mathcal{D}$ incluyendo si la secuencia es una frontera interna o externa de una cara.
4. Construir la gráfica $\mathcal{G}$ cuyos nodos corresponden con las secuencias continuas de semibordes y cuyas aristas conectan cada nodo (de frontera interna) con el nodo cuya arista se encuentra inmediatamente a la izquierda de la secuencia interna (este borde se guardó en el segundo paso).
5. Identificar cada componente conexa de $\mathcal{G}$ con una cara.  Completar la información de la tabla de semibordes con la de las nuevas caras.

```python
def Ulist(D1, D2):
    nv = len(D1['verts'])
    nf = len(D1['caras'])
    # Cambiar referencias en DF de vertices
    # Cambiar indices y columna 'b' 
    v3 = D2['verts'].copy(deep=True)
    veri = []
    b = []
    for i, row in v3.iterrows():
        # Indice
        veri.append('v' + str(int(list(i)[1]) + nv ))
        # Borde adyacente
        br = 'b' + str(int(list(row['b'])[1]) + nv ) + str(int(list(row['b'])[2]) + nv )
        b.append(br)
    v3.index = veri
    v3.b = b
    # Cambiar referencias en DF de Bordes
    b3 = D2['bordes'].copy(deep=True)
    bori = []
    for i, row in b3.iterrows():
        # Indice y gemelo
        n = list(i)
        m = list(n[:])
        n[1] = str(int(n[1]) + nv)
        n[2] = str(int(n[2]) + nv)
        bori.append(''.join(n))
        m[1] = n[2]; m[2] = n[1]
        b3.loc[i]['twin'] = ''.join(m)
        # next
        nxt = 'b' + str(int(list(row['next'])[1]) + nv) + str(int(list(row['next'])[2]) + nv)
        b3.loc[i]['next'] = nxt
        # Prev
        prv = 'b' + str(int(list(row['prev'])[1]) + nv) + str(int(list(row['prev'])[2]) + nv)
        b3.loc[i]['prev'] = prv
        # Vertice Origen
        b3.loc[i]['vert_or'] = 'v' + str(int(list(row['vert_or'])[1]) + nv)
        # Caras
        b3.loc[i]['f'] = 'f' + str(int(list(row['f'])[1]) + nf)
    b3.index = bori
    # Cambiar referencias de Caras
    c3 = m2['caras'].copy(deep=True)
    ci = []
    for i, row in c3.iterrows():
        # Indice
        ci.append( 'f' + str( int(list(i)[1]) + nf ))
        # Bordes Exteriores
        if row['b_ext'] != None:
            b1 = str(int(list(row['b_ext'])[1]) + nv)
            b2 = str(int(list(row['b_ext'])[2]) + nv)
            be = 'b' + b1 + b2
            c3.loc[i]['b_ext'] = be
        # Bordes Interiores
        if row['b_int']:
            ls = row['b_int']
            nl = []
            for j, co in enumerate(ls):
                nl.append('b' + str(int(list(co)[1]) + nv) +  str(int(list(co)[2]) + nv ))
            c3.loc[i]['b_int'] = nl
    c3.index = ci
    # Unir Listas
    v = pd.concat([D1['verts'], v3])
    b = pd.concat([D1['bordes'], b3])
    c = pd.concat([D1['caras'], c3])
    return (v, b, c)
```

```python
def calc_segs(v,b):
    segs = pd.DataFrame(columns=["x1", "y1", "x2", "y2", "br"])
    segs
    twins = []
    for i, br in b.iterrows():
        if not i in twins:
            x1 = v.loc[br['vert_or']]['x']
            y1 = v.loc[br['vert_or']]['y']
            x2 = v.loc[b.loc[br['twin']]['vert_or']]['x']
            y2 = v.loc[b.loc[br['twin']]['vert_or']]['y']
            dic = {'x1':x1, 'y1':y1, 'x2':x2, 'y2':y2, 'br':i}
            segs = segs.append(dic, ignore_index=True, sort=False)
            twins.append(br['twin'])
    # Definir puntos superiores e inferiores
    segs["x_inf"] = np.where(segs["y1"]>segs["y2"], segs["x2"], segs["x1"])
    segs["y_inf"] = np.where(segs["y1"]>segs["y2"], segs["y2"], segs["y1"])
    segs["x_sup"] = np.where(segs["y1"]>segs["y2"], segs["x1"], segs["x2"])
    segs["y_sup"] = np.where(segs["y1"]>segs["y2"], segs["y1"], segs["y2"])
    # Definir Dominio del segmento
    segs["domin"] = np.where(segs["x1"]>segs["x2"], segs["x2"], segs["x1"] ) 
    segs["domax"] = np.where(segs["x1"]>segs["x2"], segs["x1"], segs["x2"] )
    # Calcular la pendiente e intersección en el eje y
    segs["m"] = (segs["y_sup"]-segs["y_inf"])/(segs["x_sup"]-segs["x_inf"])
    segs["b"] = segs["y_sup"]-segs["m"]*segs["x_sup"]
    # Crear DF de Eventos
    eventos_sup = segs[["x_sup", "y_sup"]]
    eventos_sup.columns = ["x", "y"]
    eventos_sup["i"] = pd.Series(range(len(segs)))
    eventos_inf = segs[["x_inf", "y_inf"]]
    eventos_inf.columns = ["x", "y"]
    eventos_inf["i"] = pd.Series(range(len(segs)))
    eventos = pd.concat([eventos_sup, eventos_inf])
    eventos = eventos.sort_values(by=["y"], ascending=False)
    events = eventos.sort_values(by=["x"], ascending=True)
    events
    return (segs, eventos, events)
```

```python
def cuadrante(angle, v, b, inte):
    cua = []
    com = []
    ant = []
    sg = []
    for i, br in angle.iterrows():
        if (inte['x'] < v.loc[br['vert']]['x'] ):
            if (inte['y'] < v.loc[br['vert']]['y']):
                cua.append('I') 
                com.append(0)
            else:
                cua.append('IV') 
                com.append(360)
        else:
            if (inte['y'] < v.loc[br['vert']]['y']):
                cua.append('II') 
            else:
                cua.append('III')
            com.append(180)
        ant.append(b.loc[br['br']]['prev'])
        sg.append(b.loc[br['br']]['next'])
    angle['sig'] = sg
    angle['ant'] = ant
    angle['Cua'] = cua
    angle['com'] = com
    cua = []
    for i, ver in angle.iterrows():
        x = inte['x'] - v.loc[ver['vert']]['x']
        y = inte['y'] - v.loc[ver['vert']]['y']
        cua.append(math.degrees(np.arctan(y/x))+ ver['com'])
    angle['ang'] = cua
    angle = angle.sort_values(by=['ang'], ascending=True)
    angle.reset_index(drop=True, inplace=True)
    return angle
```

```python
def conrefer(angle, inte, v, b, sg):
    rm = []
    tw = []
    for i, ev in angle.iterrows():
        rm.append('b' + str(ev['vert']).split('v')[1] + str(ev['int']))
        tw.append('b' + str(ev['int']) + str(ev['vert']).split('v')[1])
    angle['rem'] = rm
    angle['twr'] = tw
    index = []
    agls = np.array(angle.ang, dtype=float)
    d = {'b':angle.loc[0]['twr'], 'x':inte['x'], 'y':inte['y']}
    ver = pd.DataFrame(data=d, index = ['v' + str(angle.loc[0]['int'])], columns=['b', 'x', 'y'])
    bor = pd.DataFrame(columns=['f', 'next', 'prev', 'twin', 'vert_or'])
    for i, ev in angle.iterrows():
        aux = pd.DataFrame(columns=['f', 'next', 'prev', 'twin', 'vert_or'])
        # Gemelo - Nuevo
        gn = b.loc[ev['br']]['twin']
        pt = list(angle[ angle.ang == agls[ list(agls).index( float(ev['ang'])) - 1 ] ]['rem'])[0]
        nu = list(agls).index( float(ev['ang']))
        nu = 0 if (nu == len(agls)-1) else nu + 1
        pn = list(angle[angle.ang == agls[nu]]['rem'])[0]
        # Remplazo del existente
        ea = ev['ant']
        en = list(angle[ angle.ang == agls[ list(agls).index( float(ev['ang'])) - 1 ] ]['twr'])[0]
        # Compisición de DF
        ind=[ str(ev['twr']), str(ev['rem'])]
        f = ['f', 'f']; ne = [ b.loc[gn]['next'], en]; prv = [pn, ea]
        vert = ['v' + str(ev['int']),  str(ev['vert'])]  
        d = { 'f': f, 'next': ne, 'prev': prv, 'twin': list(reversed(ind)), 'vert_or': vert }
        aux = pd.DataFrame(d)
        aux.index= ind
        b.loc[ev['ant']]['next'] = ev['rem']
        b.loc[ev['sig']]['prev'] = str(list(angle[angle.br == b.loc[ev['br']]['twin']]['twr'])[0])
        bor = bor.append(aux, ignore_index=True)
        index.extend(aux.index)
        bor.index = index
    dele = list(angle.br)
    chn = list(angle.rem)
    i = 0
    for l in dele:
        if l in list(sg.br):
            sg['br'] = sg['br'].replace( l , chn[i])
        b.drop([l], inplace=True)
        i +=1
    i = 0
    for l in dele:
        if l in list(v.b):
            v['b'] = v['b'].replace( l , chn[i])
        i +=1
    v = pd.concat([v, ver])
    b = pd.concat([b, bor])
    return  v, b, sg
```

```python
def defineverts(inte, v, bd, b, sg, segs, n):
    ba1 = segs.loc[inte['seg1']]['br']
    ba2 = segs.loc[inte['seg2']]['br']
    bn1 = sg.loc[inte['seg1']]['br']
    bn2 = sg.loc[inte['seg2']]['br']
    ve = []
    br = []
    ve, br = verseg(ba1, bn1, ve, br, inte, v, bd, b, sg, segs, n, True)
    ve, br = verseg(ba2, bn2, ve, br, inte, v, bd, b, sg, segs, n, False)
    return ve, br 
```

```python
def verseg(ba, bn,ve, br, inte, v, bd, b, sg, segs, n, ver):
    if (ba != bn ): #
        v1 = str(bd.loc[bn]['vert_or']) # 
        v2 = str(bd.loc[bd.loc[bn]['twin']]['vert_or']) #
        if ( n < int(v1.split('v')[1]) ):
            ve.append(v1)
            vn1 = str(b.loc[ba]['vert_or']) #
            vn2 = str(b.loc[b.loc[ba]['twin']]['vert_or']) #
            cl = v.loc[v1]['x'] - inte['x']
            cl1 = v.loc[vn1]['x'] - inte['x']
            cl2 = v.loc[vn2]['x'] - inte['x']
            (vn1, vn2) = (vn1, vn2) if (cl1 < cl2) else (vn2, vn1)
            (cl1, cl2) = (cl1, cl2) if (cl1 < cl2) else (cl2, cl1)
            if (cl > 0):
                ve.append(vn1)
                br.append('b' + str(v1.split('v')[1]) + str(vn1.split('v')[1]))
                br.append('b' + str(vn1.split('v')[1]) + str(v1.split('v')[1]))
            else:
                ve.append(vn2)    
                br.append('b' + str(v1.split('v')[1]) + str(vn2.split('v')[1]))
                br.append('b' + str(vn2.split('v')[1]) + str(v1.split('v')[1]))
        if (n < int(v2.split('v')[1])):
            ve.append(v2)
            vn1 = str(b.loc[ba]['vert_or'])
            vn2 = str(b.loc[b.loc[ba]['twin']]['vert_or'])
            cl = v.loc[v2]['x'] - inte['x']
            cl1 = v.loc[vn1]['x'] - inte['x']
            cl2 = v.loc[vn2]['x'] - inte['x']
            (vn1, vn2) = (vn1, vn2) if (cl1 < cl2) else (vn2, vn1)
            (cl1, cl2) = (cl1, cl2) if (cl1 < cl2) else (cl2, cl1)
            if (cl > 0):
                ve.append(vn1)
                br.append('b' + str(v2.split('v')[1]) + str(vn1.split('v')[1]))
                br.append('b' + str(vn1.split('v')[1]) + str(v2.split('v')[1]))
            else:
                ve.append(vn2)    
                br.append('b' + str(v2.split('v')[1]) + str(vn2.split('v')[1]))
                br.append('b' + str(vn2.split('v')[1]) + str(v2.split('v')[1]))
    else: 
        ve.append(str(b.loc[bn]['vert_or']))
        ve.append(str(b.loc[b.loc[bn]['twin']]['vert_or']))
        if ver:
            br.append(sg.loc[inte['seg1']]['br'])
            br.append(b.loc[sg.loc[inte['seg1']]['br']]['twin'])
        else:
            br.append(sg.loc[inte['seg2']]['br'])
            br.append(b.loc[sg.loc[inte['seg2']]['br']]['twin'])
    return ve, br
```

```python
def ManejarInterseccion(v, b, inter, segs):
    bd = b.copy(deep=True)
    vd = v.copy(deep=True)
    sg = segs.copy(deep=True)
    c =len(v) + 1    
    for i, it in inter.iterrows():
        n = (np.ones(4, dtype=int)) * c
        bor = [sg.loc[it['seg1']]['br'], sg.loc[it['seg2']]['br'] ]
        ve, br = defineverts(inter.loc[i], vd, bd, b, sg, segs, len(v))
        d = {'vert':ve, 'br': br, 'int':n }
        angle = pd.DataFrame(data=d, columns=['vert', 'br', 'int'])
        angle = cuadrante(angle, vd, bd, inter.loc[i])
        vd, bd, sg = conrefer( angle , inter.loc[i], vd ,bd, sg)
        c +=1
    return vd, bd, sg
```

### Caras

```python
def definecaras(vd, bd):
    est = []
    ls =[]
    ver = []
    for i, ev in bd.iterrows():
        w = []
        ve = []
        if not i in est:
            e = i
            w.append(e)
            ve.append(bd.loc[e]['vert_or'])
            est.append(e)
            e1 = bd.loc[e]['next']
            while (e1 != e):
                w.append(e1)
                ve.append(bd.loc[e1]['vert_or'])
                est.append(e1)
                e1 = bd.loc[e1]['next']
            ver.append(ve)
            ls.append(w)
    return { 'lb':ls, 'lv': ver }
```

```python
def clasfront(ls, vd, bd):
    bod = []; tfront = []
    fdf = pd.DataFrame(columns=['b', 't'])
    for i, li in enumerate(ls['lb']):
        x=[]
        df = pd.DataFrame(columns=['v', 'b', 'x'])
        for j, l in enumerate(li):
            x.append(vd.loc[ls['lv'][i][j]]['x'])
        df.v = ls['lv'][i]
        df.x = x
        df.b = ls['lb'][i]
        df = df.sort_values(by=['x'], ascending= True)
        df.reset_index(drop=True, inplace=True)
        vi = bd.loc[bd.loc[df.loc[0]['b']]['prev']]['vert_or']
        vf = bd.loc[bd.loc[df.loc[0]['b']]['twin']]['vert_or']
        if (vd.loc[vi]['y'] < vd.loc[vf]['y']):
            bod.append(li[0])
            tfront.append(1)
        else:
            bod.append(li[0])
            tfront.append(0)
    fdf.b = bod
    fdf.t = tfront
    return fdf
```

```python
def verizq(ls, vd, bd):
    df = pd.DataFrame(columns=['b','v', 'x'])
    ve = []
    x = []
    for i, l in enumerate(ls):
        ve.append(bd.loc[l]['vert_or'])
        x.append(vd.loc[bd.loc[l]['vert_or']]['x'])
    df.b = ls
    df.v = ve
    df.x = x
    df = df.sort_values(by=['x'], ascending= True)
    df.reset_index(drop=True, inplace=True)
    return df.loc[0]['v']
```

```python
def consgrafo(df, ls, vd, bd):
    nodos = pd.DataFrame(columns=['b','f', 's', 'i'])
    aristas = pd.DataFrame(columns=['b', 'f', 's']) 
    internos = pd.DataFrame(columns=['b', 's', 'i'])
    j = 1
    f = []; sec = []; iz = []
    for i, ev in df[df.t==0].iterrows():
        f.append('f' + str(j))
        sec.append(ls['lb'][i])
        iz.append(verizq(ls['lb'][i],vd,bd))
        j+=1
    nodos.b = df[df.t==0].b
    nodos.f=f
    nodos.s = sec
    nodos.i = iz
    # Crear eventos de Fronteras Internas
    br = []; sec = []; iz = []
    for i, l in df[df.t==1].iterrows():
        br.append(l['b'])
        sec.append(ls['lb'][i])
        iz.append(verizq(ls['lb'][i],vd,bd))
    internos.b=br
    internos.s=sec
    internos.i=iz
    for i, ev in internos.iterrows():
        sam = vd[ vd.x < vd.loc[ev['i']]['x'] ].sort_values(by=['x'], ascending=False)
    if not sam.empty:
        ver = True
        for j, e in sam.iterrows(): 
            if ver:
                for k, l in enumerate(lista_semibordes_vertice(e.index, vd, bd)):
                    if ver:
                        y1 = vd.loc[bd.loc[l]['vert_or']]['y'] 
                        y2 = vd.loc[bd.loc[bd.loc[l]['twin']]['vert_or']]['y']
                        if bd.loc[bd.loc[l]['twin']]['vert_or'] != ev['i']:
                            k1 = np.logical_and(y1 < vd.loc[ev['i']]['y'] , y2 < vd.loc[ev['i']]['y'])
                            k2 = np.logical_and(y1 > vd.loc[ev['i']]['y'] , y2 > vd.loc[ev['i']]['y'])
                            k = np.logical_or(k1, k2)
                            if not k:
                                if y1 < y2:
                                    el = bd.loc[l]['twin']
                                else:
                                    el = l
                            for c, eve in nodos.iterrows():
                                if el in list(eve['s']):
                                    ari = aristas[aristas.b==el]
                                    if ari.empty:
                                        d = { 'b':ev['b'], 'f':eve['f'], 's': ev['s']}
                                        aristas = aristas.append(d, ignore_index=True, sort=False)
                                    else:
                                        d = { 'b':ev['b'], 'f':list(ari.f)[0], 's': ev['s']}
                                        aristas = aristas.append(d, ignore_index=True, sort=False)
                                    ver = False
    else:
        n= str(len(nodos)+1)
        d = { 'b':None , 'f': 'f' +  n, 's':ev['s'], 'i':ev['i'] }
        nodos = nodos.append(d, ignore_index=True, sort=False)
        d = { 'b':ev['b'], 'f': 'f' + n, 's': ev['s']}
        aristas = aristas.append(d, ignore_index=True, sort=False)
    return {'nodos':nodos, 'aristas':aristas}
```

```python
def refcar(G, b):
    bc = b.copy(deep=True)
    caras = pd.DataFrame(columns=['b_ext', 'b_int'])
    no = G['nodos']
    ari = G['aristas']
    frin=[]
    for i, eve in no.iterrows():
        ls = list(eve['s'])
        for j, ev in enumerate(ls):
            bc.loc[ev]['f'] = str(eve['f'])
        inter = []
        q = ari[ari.f==str(eve['f'])]
        for k, e in q.iterrows():
            ls = list(e['s'])
            for j, ev in enumerate(ls):
                bc.loc[ev]['f'] = str(e['f'])
            inter.append(e['b'])
        frin.append(inter)
    caras.b_ext = list(no.b)
    caras.b_int = frin
    caras.index = list(no.f)
    return bc, caras
```

```python
def dibuja_frontera(lb, v, b):
    for l in lb:
        x1 = v.loc[b.loc[l]['vert_or']]['x']
        x2 = v.loc[b.loc[b.loc[l]['twin']]['vert_or']]['x']
        y1 = v.loc[b.loc[l]['vert_or']]['y']
        y2 = v.loc[b.loc[b.loc[l]['twin']]['vert_or']]['y']
        plt.plot([x1, x2], [y1, y2], c='r')
        plt.text( float(x1+x2)/2, float(y1+y2)/2, str(l), fontsize=10)
```

```python
def SOBREPOSICION_MAPAS(D1, D2):
    v, b, c = Ulist(m1, m2)
    segs, eventos, events = calc_segs(v, b)
    inter = Barrido_de_Linea_Abscisa(segs, eventos)
    inter = inter.append(Barrido_de_Linea_Ordenada(segs, events), sort=False)
    inter = inter.drop_duplicates(subset=['x'])
    inter.reset_index(drop=True, inplace=True)
    inter
    vc, bc, sg =  ManejarInterseccion(v, b, inter, segs)
    ls = definecaras(vc, bc)
    df = clasfront(ls, vc, bc)
    G = consgrafo(df, ls, vc, bc)
    bc, cc = refcar(G, bc)
    return { 'verts':vc, 'bordes':bc, 'caras': cc }
```
