## Aprendizaje No Supervisado 


<p style="text-align:justify;">Aprendizaje no supervisado es un método de Aprendizaje Automático donde un modelo se ajusta a las observaciones. Se distingue del Aprendizaje supervisado por el hecho de que no hay un conocimiento a priori. En el aprendizaje no supervisado, un conjunto de datos de objetos de entrada es tratado. Así, el aprendizaje no supervisado típicamente trata los objetos de entrada como un conjunto de variables aleatorias, siendo construido un modelo de densidad para el conjunto de datos. </p>

#### Clustering

<p style="text-align:justify;">En términos básicos, el objetivo de la agrupación es encontrar diferentes grupos dentro de los elementos de los datos. Para ello, los algoritmos de agrupamiento encuentran la estructura en los datos de manera que los elementos del mismo clúster (o grupo) sean más similares entre sí que con los de clústeres diferentes.</p>

<p style="text-align:justify;">El modelo de aprendizaje automático podrá inferir que hay dos clases diferentes sin saber nada más de los datos.

Estos algoritmos de aprendizaje no supervisados tienen una gama increíblemente amplia de aplicaciones y son muy útiles para resolver problemas del mundo real como la detección de anomalías, la recomendación de sistemas, la agrupación de documentos o la búsqueda de clientes con intereses comunes basados en sus compras.</p>

#### Preprocesamiento

<p style="text-align:justify;">Una vez obtenidos los documentos en texto se debe procesar la información para eliminar datos irrelevantes en el análisis, se convirtieron todas las letras a minúsculas, se removieron todos los digitos y los signos de puntuación, después se filtraron los valores correspondientes al código ascci del abecedario, lo siguiente fue disminuir los espacios repetidos, por último aplicamos la lematización, a la cual se le conoce como un proceso lingüístico que consiste en, dada una forma flexionada (es decir, en plural, en femenino, conjugada, etc), hallar el lema correspondiente. El lema es la forma que por convenio se acepta como representante de todas las formas flexionadas de una misma palabra.</p>

```python
def lsFiles(path):
    lsDir = os.walk(path)
    lsfiles = []
    for root, dirs, files in lsDir:
        for f in files:
            namefile, extension = os.path.splitext(f)
            if(extension == ".txt"):
                lsfiles.append(namefile+extension)
    return (lsfiles)

def lsCollection(lsfiles, path):
    collec = {}
    for f in lsfiles:
        namefile = path + str(f)
        data = open(namefile).read()
        #concat = Preprocessor(data)
        collec.update({str(f).split('.')[0]:data})
    return (collec)

def Preprocessor(data):
    # 1.- Convert text to lowercase
    data = data.lower()
    # 2.- Remove numbers
    data = re.sub(r"\d+", " ", data)
    # 3.- Punctuation removal
    for punc in list(string.punctuation):
        data = data.replace(punc, " ")
    data = data.strip()
    # 4.- Remove hexa codes (0xef)
    concat = ""
    for word in data.split():
        for w in word:
            if np.logical_and(ord(w)>=97, ord(w)<=122):
                concat = concat + w
            else:
                concat = concat + " "
        concat = concat + " "
    data = re.sub(r"\ +", " ", concat)
    # 5.- White spaces removal
    data = data.strip()
    #6.- Lemmatization
    lemmatizer=WordNetLemmatizer()    
    concat = ""
    lswords=word_tokenize(data)
    for word in lswords:
        concat = concat + " " + lemmatizer.lemmatize(word)
    data = concat
    return data

def Bag_of_Words(data, lan="english"):
    # Creación de stop words con la librería nltk
    stop_words = list(stopwords.words(lan))
    #datac = Preprocesor(data)

    BW = set()
    # Agregamos cada palabra a la bolsa de palabras
    for wd in data.split(" "):
        if ((not wd in stop_words) and (len(wd)>3)):
                BW.add(wd)
    return BW

def Set_Vocabulary(collec):
    vocabulary = set()
    for key, f in collec.iteritems():
        BW = Bag_of_Words(f)
        vocabulary = vocabulary.union(BW)
    return (vocabulary)

def Inverted_Index(collec):
    inv_ind = {}
    for file in collec:
        BW = Bag_of_Words(collec[file])
        for t in BW:
            if not t in inv_ind:
                inv_ind[t] = set([file])
            else:
                inv_ind[t].add(file)
    sorted_keys = sorted(inv_ind.keys())
    return (sorted_keys, inv_ind)

# Crea una gráfica de barras de frecuencia
def Frequency_Bar(cluster):
    groups, counts = np.unique(cluster, return_counts=True)
    d = {"Group":groups, "Total":counts}
    df = pd.DataFrame(d)
    df.plot.bar(x="Group", y="Total", rot=0)
    plt.show()

# Selecciona las palabras destacadas del articulo
def Top_Words(doc, ntop=10):
    topw = df.nlargest(ntop, [doc])
    result = topw[doc]
    return result

# Construye una nube de términos
def Word_Cloud(title, words):
    wordcloud = WordCloud(width=1000, height=1000, background_color = "white", 
                          min_font_size=14).generate(" ".join(words))
    plt.figure(figsize=(4,4), facecolor=None)
    plt.imshow(wordcloud)
    #plt.title(title)
    plt.axis("off")
    plt.tight_layout(pad = 0)
    print(title)
    plt.show()
    
# Obtiene algunos términos de un representante de grupo
def Centroid_Features(i, order_centroids, n_words=50):
    terms = feature_names
    cloud = []
    for ind in order_centroids[i, :n_words]:
        cloud.append(terms[ind])
    return cloud
```

## Term Frequency - Inverse Document Frequency (tf-idf)


<p style="text-align:justify;">Frecuencia de término – Frecuencia inversa de documento (o sea, la frecuencia de ocurrencia del término en la colección de documentos), es una medida numérica que expresa cuán relevante es una palabra para un documento en una colección. Esta medida se utiliza a menudo como un factor de ponderación en la recuperación de información y la minería de texto. El valor tf-idf aumenta proporcionalmente al número de veces que una palabra aparece en el documento, pero es compensada por la frecuencia de la palabra en la colección de documentos, lo que permite manejar el hecho de que algunas palabras son generalmente más comunes que otras.

Variaciones del esquema de peso tf-idf son empleadas frecuentemente por los motores de búsqueda como herramienta fundamental para medir la relevancia de un documento dada una consulta del usuario, estableciendo así una ordenación o ranking de los mismos. Tf-idf puede utilizarse exitosamente para el filtrado de las denominadas stop-words (palabras que suelen usarse en casi todos los documentos), en diferentes campos como la clasificación y resumen de texto.</p>

<p style="text-align: center;"> $W_{ij} = t_{f_{ij}}* log(\frac{N}{d_{f_{j}}})$</p>

## Agrupación K-Medias

<p style="text-align:justify;">El algoritmo K-Means tiene como objetivo encontrar y agrupar en clases los puntos de datos que tienen una alta similitud entre ellos. En los términos del algoritmo, esta similitud se entiende como lo opuesto de la distancia entre puntos de datos. Cuanto más cerca estén los puntos de datos, más similares y con más probabilidades de pertenecer al mismo clúster serán.</p>

#### Pasos del Algoritmo

1. Primero, se necesita elegir k, el cuál representa el número de clusters que queremos que nos encuentren.
2. Segundo, el algoritmo seleccionará aleatoriamente los centroides de cada grupo.
3. Tercero, se asignará cada punto de datos al centroide más cercano (utilizando la distancia euclídea).
4. Cuarto, se calcula la inercia del conglomerado.
5. Quinto, los nuevos centroides se calcularán como la media de los puntos que pertenecen al centroide del paso anterior. En otras palabras, calculando el error cuadrático mínimo de los puntos de datos al centro de cada cluster, moviendo el centro hacia ese punto.
6. Sexto, volver al paso 3.

#### Parámetros de K-Means

- n_cluster: El número de clusters y centroides de cada grupo.
- max_iter: Número de iteraciones del algoritmo para una sola ejecución.

#### Seleccionar K

<p style="text-align:justify;">La elección del número correcto de clusters es uno de los puntos clave del algoritmo K-Means. En nuestro esta ocasión utilizamos el metodo del codo (elbow method). Dicho método funciona trazando los valores ascendentes de K frente al error total obtenido al usar esa K.</p>

\begin{align}
    \%Varianza = \frac{Varianza\ entre\ grupo}{Varianza\ Total}\\ 
\end{align}

El objetivo es encontrar la k adecuada para cada cluster donde no aumente significativamente la varianza.

```python
# Instantiate the clustering model and visualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(4,30), locate_elbow=True)

visualizer.fit(X)        # Fit the data to the visualizer
visualizer.poof() # Draw/show/poof the data
```

<p style="text-align:justify;">Aplicamos nuestro modelo de agrupación de <i>K-Medias</i> a nuestro conjunto de datos. Y visualizamos la asignación de cada grupo que se le dio a cada documento de la matriz. La mejor <i>k</i> obtenida visualizando la gráfica es $k = 9$ porque la disminución de errores es más clara con ese valor para k.</p>

```python
K = 9
kmeans = KMeans(n_clusters=K, max_iter=100)
kmeans.fit(X)
order_centroids = kmeans.cluster_centers_.argsort()[:,::-1]
cluster = kmeans.predict(X)
print(cluster)
```

<p style="text-align:justify;"> Se visualiza por medio de un gráfico la cantidad de documentos por cada grupo obtenido de nuestro modelo de agrupación por <i>K-Medias</i>:</p>

```python
Frequency_Bar(cluster)
```
