<p  style="text-align:justify" >Realizamos un resumen de las características numéricas para conocer sus medidas estadísticas como la media y la desviación estándar, así como los valores máximos y mínimos para cada variable. Otros datos obtenidos son los valores de los percentiles en la distribución de los datos.</p>

<p  style="text-align:justify" >Construimos funciones para visualizar una gráfica de pares con las características numéricas del conjunto de datos, la gráficas de pares se divide de la siguiente forma sobre la matriz de gráficas:
<ul>
    <li>Diagonal Principal: Histograma de frecuencias. </li>
    <li>Triangular Superior: Coeficiente de Correlación. </li>
    <li>Triangular Inferior: Diagrama de Dispersión. </li>
</ul>
</p>

```python
# Function to calculate correlation coefficient between two arrays
def corr(x, y, **kwargs): 
    # Calculate the value
    coef = np.corrcoef(x, y)[0][1]
    # Make the label
    label = r'$\rho$ = ' + str(round(coef, 2))
    # Add the label to the plot
    ax = plt.gca()
    ax.annotate(label, xy = (0.3, 0.5), size = 20, xycoords = ax.transAxes)

def Plot_Pairs(df, col):
    lk = list(df["Alert"].unique())
    c = [ "r", "g", "b"]
    colours = [c[lk.index(row["Alert"])] for i, row in df.iterrows() ]
    df0 = df.copy(deep=True)
    df0.drop(["Alert"], axis='columns', inplace=True)
    # Create a pair grid instance
    grid = sb.PairGrid(data= df0, vars = col, size = 4)
    # Map the plots to the locations
    grid = grid.map_lower(plt.scatter, color = colours, alpha=0.25)
    grid = grid.map_upper(corr)
    grid = grid.map_diag(plt.hist, bins = 10, edgecolor =  'k', color = 'blue')
    
# Columnas a graficar
columns = ['IQR', 'Kurto', 'Entro', 'Skewn', 'Umbral' ]
```

<p  style="text-align:justify" >Ahora visualizamos la gŕafica de pares sobre el conjunto de datos de Karen App.</p>

```python
Plot_Pairs(df, columns)
```

### Escalar las características categóricas
<br>
<p  style="text-align:justify" >Ahora realizamos una transformación a las variables categóricas con el método <i>get_dummies</i> de <i>pandas</i> que nos permite binarizar las variables categóricas para evitar durante su procesamiento influencias no deseadas por el mal escalamiento de los datos.</p>
<p  style="text-align:justify" >Escalamos las características del conjunto de datos para visualizar que tanto cambia con respecto al conjunto de datos original.</p>

```python
# Variables categoricas transformadas
df_dummie = pd.get_dummies(df, drop_first=True)
# Dataframe dummie escalado
df_ds = scale(df_dummie)
df_ds=pd.DataFrame(df_ds, columns=df_dummie.columns.values)
df_ds.head()
```

<p  style="text-align:justify" >Resumen de la distribución de los datos escalados.</p>

```python
df_ds.describe()
```

## Separar los datos en conjuntos de entrenamiento y prueba

<p  style="text-align:justify" >A continuación realizamos la separación del conjunto de datos en dos, el conjunto de entrenamiento de encarga de la construcción de los clasificadores, y el conjunto de prueba es para evaluar la precisión de nuestro modelo de clasificación. Nuestro conjunto de entrenamiento contiene 75% de la información y el conjunto de prueba 25% de la información, con una semilla de valor 100 para siempre obtener los mismos resultados en la separación de los datos.</p>

```python
# Dataframe dummie copia
df_dc = df_dummie.copy(deep=True)
y = df_dc["Alert"]
df_dc = df_dc.drop(['Alert'], axis=1)
X = df_dc
# Separación de conjuntos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=100)
print "Cantidad de datos en el conjunto de Entrenamiento: " + str(X_train.count()[0])
print "Cantidad de datos en el conjunto de Prueba: " + str(X_test.count()[0])
```

## Proyectar los datos en dos dimensiones

<p  style="text-align:justify" >Función para graficar las características del conjunto de datos en un espacio de dos dimensiones.</p>

```python
def Plot_Classes(df, projection):
    # Separar clases
    clase1 = df[df.Alert]
    clase2 = df[-df.Alert]
    # Title
    title = ["Datos sin estandarizar", "Datos estandarizados"] 
    # figure
    fig = plt.figure(figsize=(18, 5))

    for i, data in enumerate(projection):
        axis = fig.add_subplot(1, len(title),i+1)
        axis.plot(data[clase1.index, 0], data[clase1.index, 1], "r.", alpha=0.25)
        axis.plot(data[clase2.index, 0], data[clase2.index, 1], "b.", alpha=0.25)   
        axis.set_title(title[i])
    plt.show()
```

<p  style="text-align:justify" >A partir de las siguientes gŕaficas podemos comparar los datos estandarizados y sin estandarizar.</p>

```python
pca = PCA(n_components=2) 
# Datos sin estandarizar
pca.fit(df_dummie)
df_projection = pca.transform(df_dummie)
# Datos Estandarizados
pca.fit(df_ds)
df_projection_s = pca.transform(df_ds)
# Graficar datos
Plot_Classes(df, [df_projection, df_projection_s])
```

# K - Nearest Neighbors

<p  style="text-align:justify" >Aplicamos el método de vecinos más cercanos por medio del clasificador <i>KNeighborsClassifier()</i> obtenido de los módulos de <i>sklearn</i>, construimos un <i>Pipeline</i>, primero con el comando de escalamiento de datos para posteriormente aplicar el clasificador, definimos nuestros parámetros del clasificador para realizar una búsqueda exhaustiva con validación cruzada, con un valor de <i>k</i>=10, para realizar la búsqueda utilizamos el comando <i>GridSearchCV()</i> proveniente de los módulos de sklearn para encontrar los mejores parámetros para nuestro modelo de clasificación.</p>
<p  style="text-align:justify" >Después de realizar la validación de parámetros debemos entrenar nuestro modelo con el conjunto de datos de entrenamiento.</p>

```python
steps = [('scaler', StandardScaler()), 
         ('knn', KNeighborsClassifier())]
pipeline = Pipeline(steps)

parameters = {'knn__n_neighbors': list(range(1,31)),
              'knn__weights': ["uniform", "distance"]}

knn_cv = GridSearchCV(pipeline, parameters, cv=10)
knn_cv.fit(X_train, y_train)
```

```python
print("Mejores Parámetros del modelo: {}".format(knn_cv.best_params_))
```

#### Métricas de Evaluación de Clasificación (Confusion Matrix, Accuracy, Precision, Recall, F1, ROC y AUC)

<ul>
 
<li><p  style="text-align:justify" >Una matriz de confusión es una tabla que describe el desempeño de un modelo de clasificación en un conjunto de datos de prueba cuyos valores verdaderos son conocidos. Una matriz de confusión es altamente interpretativa y puede ser usada para estimar un número de otras métricas.</p></li>

<li><p  style="text-align:justify" >La exactitud de la clasificación es la relación entre las predicciones correctas y el número total de predicciones. O más simplemente, con qué frecuencia es correcto el clasificador.</p></li>

<li><p  style="text-align:justify" >La precisión es la relación entre las predicciones correctas y el número total de predicciones correctas previstas. Esto mide la precisión del clasificador a la hora de predecir casos positivos.</p></li>

<li><p  style="text-align:justify" >La sensibilidad también es llamada en inglés recall, es la relación entre las predicciones positivas correctas y el número total de predicciones positivas. O más simplemente, cuán sensible es el clasificador para detectar instancias positivas. Esto también se conoce como la tasa verdadera positiva.</p></li>

<li><p  style="text-align:justify" >El puntaje F1 es la medida armónica de la memoria y la precisión, con una puntuación más alta como mejor modelo.</p></li>

<li><p  style="text-align:justify" >Una curva ROC (curva de característica operativa del recepto) es un gráfico que muestra el rendimiento de un modelo de clasificación en todos los umbrales de clasificación. Esta curva representa dos parámetros:</p>
    <ul>
    <li>Tasa de verdaderos positivos.</li>
    <li>Tasa de falsos positivos.</li></ul></li>
    
<li><p  style="text-align:justify" >AUC significa "área bajo la curva ROC". Esto significa que el AUC mide toda el área bidimensional por debajo de la curva ROC completa. El AUC proporciona una medición agregada del rendimiento en todos los umbrales de clasificación posibles. Una forma de interpretar el AUC es como la probabilidad de que el modelo clasifique un ejemplo positivo aleatorio más alto que un ejemplo negativo aleatorio. </p> </li>
</ul>


<p  style="text-align:justify" >Función para visualizar la matriz de confusión.</p>

```python
# funcion optenida de: https://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html
def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion matrix', cmap=plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    #plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
    
def Plot_ROC_Curve(y_test, y_pred_prob):
    # Generate ROC curve values: fpr, tpr, thresholds
    fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
    # Plot ROC curve
    plt.plot([0, 1], [0, 1], 'k--')
    plt.plot(fpr, tpr,  marker='.')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.show()
```

```python
# Predicted classes
y_pred = knn_cv.predict(X_test)
# Compute predicted probabilities: y_pred_prob
y_pred_prob = knn_cv.predict_proba(X_test)[:,1]

# Compute and print metrics

# Compute Confusion matrix
cm = confusion_matrix(y_test, y_pred)
plot_confusion_matrix(cm,["1","0"])

print("Métricas del modelo: \n")

knn_accu = accuracy_score(y_test, y_pred)
knn_pre = precision_score(y_test, y_pred)
knn_rec = recall_score(y_test, y_pred)
knn_f1 = f1_score(y_test, y_pred)
print("Exactitud: {:.0f}%".format(100*knn_accu))
print("Precisión: {:.0f}%".format(100*knn_pre))
print("Sensibilidad: {:.0f}%".format(100*knn_rec))
print("Puntaje F1: {:.0f}%".format(100*knn_f1))

print("\nAlternativa con Classification_Report: \n")
# Compute Classification Report
print classification_report(y_test, y_pred)
# Plot ROC Curve
Plot_ROC_Curve(y_test, y_pred_prob)
# AUC 
knn_auc = roc_auc_score(y_test, y_pred_prob)
print("AUC: {:.0f}%".format(100*knn_auc))
```

# Neural Network

```python
print("Mejores Parámetros del modelo: {}".format(MLPC_cv.best_params_))
```
