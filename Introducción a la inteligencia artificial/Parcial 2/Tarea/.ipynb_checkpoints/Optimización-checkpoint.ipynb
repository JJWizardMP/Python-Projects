<p style="text-align:justify">
La optimización es fundamental para cualquier problema relacionado con la toma en decisiones de diversas áreas. La aplicación de los métodos de optimización reside en la posibilidad de determinar la solución óptima sin tener que probar todos los casos posibles.
</p>
<p style="text-align:justify">
La optimización de funciones comienza con un conjunto de variables independientes o parámetros, y a menudo incluye condiciones o restricciones que definen los valores aceptables de estas variables. Tales restricciones se denominan las limitaciones del problema. Otro componente esencial de un problema de optimización es una medida única denominada función objetivo, la cual depende también de las variables independientes. La solución al problema de optimización estará dada por un conjunto de valores permitidos para las variables independientes, de acuerdo a las limitaciones del caso particular en los que la función objetivo asume un valor óptimo. En términos matemáticos, la optimización implica normalmente maximizar o minimizar la función de costo.</p>

### Funciones de costo

<ul>
    <li><p style="text-align:justify">
<strong>De Jong's</strong>: Conocida como la primera función de De Jong's, es una de las pruebas de referencia más simples. La función es continua, convexa y unimodal.</p></li>
    <li> <p style="text-align:justify">
<strong>Rosenbrock's Valley</strong>: El valle de Rosenbrock es un problema clásico de optimización, también conocido como función banana o la segunda función de De Jong. El óptimo global se encuentra dentro de un valle plano largo, estrecho y de forma parabólica. Encontrar el valle es trivial, sin embargo, la convergencia al óptimo global es difícil y, por lo tanto, este problema se ha utilizado con frecuencia para probar el rendimiento de los algoritmos de optimización.</p></li>
        <li> <p style="text-align:justify">
<strong>Rastrigin's</strong>: La función de Rastrigin se basa en la función de De Jong con la adición de la modulación del coseno para producir mínimos locales frecuentes. Por lo tanto, la función de prueba es altamente multimodal. Sin embargo, la ubicación de los mínimos se distribuye de forma regular.</p></li>
            <li> <p style="text-align:justify">
<strong>Rotated Hyper-Ellipsoid</strong>: Una extensión de la función de costo de <i>axis parallel hyper-ellipsoid</i>, el cual es la función <i>Schewefelp's</i> con respecto a los ejes de coordenadas, esta función produce hiperelipsoides rotados. La función es continua, convexa y unimodal.</p></li>
            <li> <p style="text-align:justify">
<strong>Griewangk's</strong>: La función de Griewangk es similar a la función de Rastrigin. Tiene muchos mínimos locales generalizados distribuidos de forma regular.</p></li>
</ul>


```python
def Select_Function(x, y, func = None):
    z = 0
    if func == "Rastrigin's":
        z = 20+x**2 + y**2 - 10*np.cos(2*np.pi*x) - 10*np.cos(2*np.pi*y)
    elif func == "De Jong's":
        z = x**2 + y**2
    elif func == "Rosenbrock's Valley":
        z = 100*(y-x**2)**2 + (1-x)**2
    elif func == "Rotated Hyper-Ellipsoid":
        z = x**2 + (x**2 + y**2)
    elif func == "Griewangk's":
        z = ((x**2 + y**2)/400) - np.cos(x)*np.cos(y/np.sqrt(2)) + 1
    return z
```

```python
def Plot_Functions(funcls):
    #Gráficas de curvas de nivel
    X = np.linspace(-5,5,100)
    Y = np.linspace(-5,5,100)
    X, Y = np.meshgrid(X, Y)
    fig = plt.figure(figsize=(18, 5))
    for i, f in enumerate(funcls):
        Z = Select_Function(X,Y,f)
        axis = fig.add_subplot(2, 3,i+1, projection="3d")
        axis.plot_surface(X,Y,Z, cmap= cm.Spectral, rstride=1, cstride=1, linewidth=0, antialiased=True, 
                          shade = True, alpha=1)
        axis.contour3D(X, Y, Z, 50)
        axis.set_title(f)
    plt.show()
```

# Algoritmos Genéticos


<p style="text-align:justify">
Un algoritmo genético consiste en una función matemática o una rutina de software que toma como entradas a los ejemplares y retorna como salidas cuales de ellos deben generar descendencia para la nueva generación. estos algoritmos funcionan entre el conjunto de soluciones de un problema llamado fenotipo, y el conjunto de individuos de una población natural, codificando la información de cada solución en una cadena, generalmente binaria, llamada cromosoma. Los símbolos que forman la cadena son llamados genes. Cuando la representación de los cromosomas se hace con cadenas de dígitos binarios se le conoce como genotipo. Los cromosomas evolucionan a través de iteraciones, llamadas generaciones. En cada generación, los cromosomas son evaluados usando alguna medida de aptitud. Las siguientes generaciones (nuevos cromosomas), son generadas aplicando los operadores genéticos repetidamente, siendo estos los operadores de selección, cruzamiento, mutación y reemplazo. 
</p>

Ventajas y desventajas:

<ul>
    <li><p style="text-align:justify">Operan de forma simultánea con varias soluciones, en vez de trabajar de forma secuencial como las técnicas tradicionales.</p></li> 
    <li><p style="text-align:justify">
    Cuando se usan para problemas de optimización maximizar una función objetivo- resultan menos afectados por los máximos locales (falsas soluciones) que las técnicas tradicionales.</p></li>
    <li><p style="text-align:justify">
        Resulta sumamente fácil ejecutarlos en las modernas arquitecturas masivamente paralelas.</p></li> 
    <li><p style="text-align:justify">
        Usan operadores probabilísticos, en vez de los típicos operadores determinísticos de las otras técnicas</p></li> 
    <li><p style="text-align:justify">
        Pueden tardar mucho en converger, o no converger en absoluto, dependiendo en cierta medida de los parámetros que se utilicen tamaño de la población, número de generaciones, etc.</p>
        <li><p style="text-align:justify">Pueden converger prematuramente debido a una serie de problemas de diversa índole. </p></li>
</ul>
    
¿Cuándo aplicar algoritmos genéticos? 
<ul>
    <li>Cuando su espacio de búsqueda esté delimitado dentro de un cierto rango.</li>
    <li><p style="text-align:justify">Debe poderse definir una función de aptitud que nos indique qué tan buena o mala es una cierta respuesta.</p></li> 
    <li><p style="text-align:justify">Las soluciones deben codificarse de una forma que resulte relativamente fácil de implementar en la computadora.</p></li>

## Representación

<p style="text-align:justify">En los algoritmos genéticos la información de un individuo se codifica mediante una cadena binaria de longitud $L$</p>

$$\color{blue}{I = B_1B_2B_3\cdots B_L}$$

Por ejemplo $I = 1001001$.
<p style="text-align:justify">
Supón que en el problema de optimización, la función de costo, $f(x)$, depende de una sola variable. Supón también que la variable está limitada a un rango de valores que va de $x_{min}$ hasta $x_{max}$. Podemos entonces asociar cada individuo con un valor de $x$ dentro del intervalo $[x_{min}, \;x_{max}]$ de la siguiente forma:</p>

$$\color{blue}{x(I) = x_{min}+D(I)\frac{x_{max}-x_{min}}{2^{N}-1}}$$

En donde $D(I)$ es el valor en base diez del número en base binaria definido en la cadena del individuo.
<p style="text-align:justify">
Por ejemplo, supón que se utiliza una cadena de dos bits para representar a los individuos y que la variable está acotada al intervalo $[1,\; 2]$. Los cuatro posibles individuos son los siguientes:</p>

\begin{align}
    I_1 &= 00\\
    I_2 &= 01\\
    I_3 &= 10\\
    I_4 &= 11\\
\end{align}

Los valores en base diez de cada uno son 

\begin{align}
    D(I_1) &= 0\\
    D(I_2) &= 1\\
    D(I_3) &= 2\\
    D(I_4) &= 3\\
\end{align}

y cada uno de ellos se mapearía a los siguientes valores reales

\begin{align}
    x(I_1) &= 1+0\frac{1}{2^2-1} = 1\\
    x(I_2) &= 1+1\frac{1}{2^2-1} = 1\frac{1}{3}\\
    x(I_3) &= 1+2\frac{1}{2^2-1} = 1\frac{2}{3}\\
    x(I_2) &= 1+3\frac{1}{2^2-1} = 2\\
\end{align}

### Problemas multivariados

Si la función de costo depende de más de una variable, entonces para cada variable se define una cadena binaria de longitud adecuada y éstas se concatenan para formar al individuo completo. Por ejemplo, supón que la función de costo depende de dos variables y  que se utilizan tres bits para representar a cada uno de ellos, el individuo $I = 010110$ tiene asociadas las subcadenas $s_1 = 010$ y $s_2=110$ con cada una de las variables. </p>

## Población inicial 

<p style="text-align:justify">Una vez que se define la representación de cada individuo se debe iniciar la <em>población</em>. Dicha población es un simple conjunto de individuos generados de forma aleatoria. Supongamos por ejemplo supón que el problema de optimización es tal que la función de costo depende de dos variables. Supón también que hemos decidido utilizar 4 bits para codificar cada variable y que la población quedará definida por 10 individuos.  
</p>
<p style="text-align:justify">
La siguiente función permite iniciar a dicha población</p>

```python
def IniciaPob(N,n_var,n_bits):
    Pob = []
    for i in range(N):
        I = ''
        for v in range(n_var):
            for b in range(n_bits):
                I = I+sample(['0','1'],1)[0]
        Pob.append(I)
    return Pob
```

<p style="text-align:justify">A cada individuo le corresponde una ubicación en el espacio bidimensional, pero para graficarlo es necesario tener una función que brinde los valores reales de las variables asociadas con cada individuo</p>

```python
def Ind2Number(I, v_min, v_max, n_vars, n_bits):        
    numb = []
    for i in range(n_vars):
        cod_v = I[i*n_bits:(i+1)*n_bits]
        d_i = int(cod_v,2)
        val = v_min[i]+d_i*(v_max[i]-v_min[i])/float(2**n_bits-1)
        numb.append(val)
    return numb        
```

```python
def Pob2Number(P):
    Pob = [ Ind2Number(p, v_min, v_max, n_var, n_bits ) for p in P ]
    return Pob
```

## Función de Aptitud

<p style="text-align:justify">A cada individuo se le asocia un valor de aptitud que refleja la calidad del individuo como solución del problema. Para los problemas de optimización la función de aptitud debe ser inversamente proporcional a la función de costo. </p>

Una posibilidad para la función de aptitud es

$$A(x_1,x_2) = \frac{1}{1+\alpha f(x_1,x_2)}$$

Siendo $\alpha$ un parámetro que permite calibrar dicha función. 

```python
def Aptitude(I, f, t):
    if t == "Gen":
        [x1,x2] = Ind2Number(I,v_min,v_max,n_var,n_bits)
    elif t == "Evo":
        [x1, x2] = I[0], I[1]
    r = Select_Function(x1, x2, f)
    return 1.0/(1.0+0.1*r)    
```

```python
def Pop_Aptitude(P, f, t):
    apt_v = [ Aptitude(I, f, t) for I in P ]
    return apt_v
```

```python
#selecciona un individuo de la población de acuerdo a su aptitud
def select(P,Aptitude):
    apt_acum = sum(Aptitude)
    r = random()*apt_acum
    s = 0
    for i,I in enumerate(P):
        s = s+Aptitude[i]
        if s >= r:
            return I
    return I
```

## Operadores Genéticos

### Cruza

<p style="text-align:justify">Por medio del operador genético de cruza es posible intercambiar la información contenida en dos individuos. Los individuos con los que se realizará la cruza deben ser seleccionados con una probabilidad proporcional a su aptitud.</p>
<p style="text-align:justify">
Existen varias opciones para realizar la cruza una vez que se seleccionaron los individuos. Una de las más utilizadas es la cruza en un solo punto. En esta estrategia</p>

1. Se selecciona aleatoriamente una posición sobre la cual se cortarán ambos individuos en dos subcadenas. La posición debe estar comprendida entre el segundo y el penúltimo índices.
2. El nuevo individuo se genera seleccionando la primera subcadena del primer individuo y concatenándola con la segunda subcadena del segundo individuo.

Por ejemplo, supón que los individos seleccionados para la cruza son los siguientes

\begin{align}
    I_1 &= \color{blue}{001}\color{red}{01101}\\
    I_2 &= \color{red}{010}\color{blue}{11010}
\end{align}

Supón que la posición seleccionada es la tercera. El nuevo individuo sería

$$I_n = \color{blue}{00111010}$$

```python
def Genetic_CrossOver(I1,I2):
    n = len(I1)
    p = randint(1,n-2)
    b1 = I1[:p]
    b2 = I2[p:]
    return b1+b2
```

<p style="text-align:justify">Cada vez que se aplica la cruza se genera un nuevo individuo y este proceso se debe repetir por cada uno de los individuos de la población original, con lo que se genera una nueva población</p>

```python
#probabilidad de cruza = 100%
def Genetic_CrossOverPop(P,Aptitude):
    CrossP = []
    for I in P:
        I1 = select(P,Aptitude)
        I2 = select(P,Aptitude)
        nI = Genetic_CrossOver(I1,I2)
        CrossP.append(nI)
    return CrossP        
```

### Mutación

<p style="text-align:justify">En la mutación perturbamos con cierta probablidad la información genética contenida en un individuo. Para esto debemos recorrer cada bit del individuo y decidir si se cambia el valor de dicho bit en base a una probabilidad definida, llamada probabilidad de mutación: $p_m$. Por ejemplo, si $p_m = 0.01$, se debe cambiar aproximadamente el valor de uno de cada cien bits.</p>

```python
def Genetic_Mutation(P,pm):
    PM = []
    cs = {'0':'1','1':'0'}
    for I in P:
        nI = ''
        for b in I:
            if random() < pm:
                nI = nI+cs[b]
            else:
                nI = nI+b
        PM.append(nI)
    return PM    
```

### Selección

<p style="text-align:justify">A través de la selección elegimos entre varios individuos aquellos que tengan el mayor valor de aptitud. Por ejemplo, si se tienen dos poblaciones, entre una pareja de individuos elegidos aleatoriamente de cada población, se selecciona el que tenga la mayor aptitud para formar parte de la nueva población</p>

```python
def SelectP(P1,P2,A1,A2):
    PS = []
    for i in range(len(P1)):
        if A1[i] > A2[i]:
            PS.append(P1[i])
        else:
            PS.append(P2[i])
    return PS
```

## Evolución

<p style="text-align:justify">La evolución en los algoritmos genéticos se realiza aplicando sucesivamente los operadores de cruza, mutación y selección a una población. En cada aplicación se genera una nueva población. La esperanza es que la evolución permita construir una secuencia de poblaciones que converja hacia la solución del problema de optimización. El número de veces que se aplica la secuencia de operadores se denomina <em>generaciones</em></p>

```python
def Genetic_Evolution(P0,func,pm,gnrtns):
    P = P0
    for g in range(gnrtns):
        AP = Pop_Aptitude(P, func, "Gen")
        PC = Genetic_CrossOverPop(P,AP)
        PM = Genetic_Mutation(PC,pm)
        AM = Pop_Aptitude(PM, func, "Gen")
        P1 = SelectP(P,PM,AP,AM)
        P = P1
    return P1        
```

# Estrategias Evolutivas
<p style="text-align:justify">
Son abstracciones basadas en la teoría de la evolución biológica y usadas para crear procedimientos de optimización o metodologías:</p>
<ul>
    <li><p style="text-align:justify">
        Son implementadas en problemas computacionales.</p></li>

        Son usadas para resolver problemas de optimización</p></li>
<p style="text-align:justify">
Un argumento general que es usado para justficar el uso de la computación evolutiva es el siguiente:</p>
<p style="text-align:justify">"La evolución ha optimizado los procesos biológicos, por lo tanto, el uso del paradigma evolutivo en la computación aplicado a ciertos problemas, podría ayudarnos a encontrar soluciones óptimas y originales."</p>
<p style="text-align:justify">La evolución es un proceso paralelo ya que las operaciones de evaluación y cambios a las numerosas especies e individuos sucede al mismo tiempo (en paralelo).</p>
<p style="text-align:justify">
    Los conceptos de la selección natural:</p>
    
<ul>
    <li><p style="text-align:justify">Limitado número de recursos.</p></li>

    El éxito depende de la aptitud del individuo (qué tan buien adaptado está el individuo en su ambiente determinado por los genes).</p></li>

## Operadores Evolutivos

### Cruza

<p style="text-align:justify">Se seleccionan <i>P</i> individuos Padres y el nuevo individuo será el centroide de los individuos padres seleccionados.
</p>
<br>
\begin{align}
    P &= 2  && I_1 = [x_1, x_2, x_3 ]\\
    n &= 3  && I_2 = [y_1, y_2, y_3 ]\\
            &&& I' = [ \frac{x_1 + y_1}{2},\frac{x_2 + y_2}{2}, \frac{x_3 + y_3}{2} ] \\
\end{align}

```python
def Evolution_CrossOver(I1, I2):
    I = [ (I1[i]+I2[i])/2 for i in range(len(I1)) ]
    return I
```

```python
def Evolution_CrossOverPop(P,Aptitude, lam):
    CrossP = []
    for i in range(lam):
        I1 = select(P,Aptitude)
        I2 = select(P,Aptitude)
        nI = Evolution_CrossOver(I1,I2)
        CrossP.append(nI)
    return CrossP        
```

### Mutación

<p style="text-align:justify">La operación de Mutación suele ser un operador de variación básica en las Estrategias Evolutivas. Es decir, es la fuente principal de la variación genética. El diseño de los operadores de mutación depende del problema.</p>
\begin{align}
    &I_1 = [x_1, x_2, ..., x_n ]\\
    & I' = [x_1 + randNorm(0,\sigma), x_2 + randNorm(0,\sigma), ..., x_n + randNorm(0,\sigma)] \\
\end{align}

```python
def Evolution_Mutation(P, sd):
    mutated = [ [p[0]+np.random.normal(loc=0, scale=sd), p[1]+np.random.normal(loc=0, scale=sd) ] for p in P ]
    return mutated
```

### Selección

<p style="text-align:justify">La selección en las Estrategias Evolutivas consiste en unir a los padres e hijos para seleccionar a los mejores &mu; individuos:</p>
\begin{align}
    (\mu + \lambda) \leftarrow Estrategia Elitista
\end{align}
<p style="text-align:justify">Una segunda estrategia, se seleccionan a los &mu; individuos de la población de hijos:</p>
\begin{align}
    (\mu, \lambda) \leftarrow Donde && \lambda > \mu
\end{align}

## Evolución

<p style="text-align:justify">La evolución en los algoritmos genéticos se realiza aplicando sucesivamente los operadores de cruza, mutación y selección a una población. En cada aplicación se genera una nueva población. La esperanza es que la evolución permita construir una secuencia de poblaciones que converja hacia la solución del problema de optimización. El número de veces que se aplica la secuencia de operadores se denomina <em>generaciones</em></p>

```python
def Evolution_Strategy(P0, func, lam, sd, gnrtns):
    P = Pob2Number(P0)
    for i in range(gnrtns):        
        AP = Pop_Aptitude(P, func, "Evo")
        PC = Evolution_CrossOverPop(P,AP, lam)
        PM = Evolution_Mutation(PC, sd)
        AM = Pop_Aptitude(PM, func, "Evo")
        P1 = SelectP(P,PM,AP,AM)
        P = P1
    return P1
```

 # Descenso de Degradiente
 
<p style="text-align:justify">El método de descenso de gradiente es un algoritmo de optimización que permite converger hacia el valor mínimo de una función a través de un proceso iterativo. Es utilizado en el aprendizaje automático para minimizar una función que mide los errores de predicción del modelo en el conjunto de datos. A esta función de error se le suele denominar función de costo e identificar con <i>J(θ)</i>, en donde θ hace referencia a los parámetros del modelo. </p>
    
<p style="text-align:justify">Para identificar el mínimo de la función el método del descenso de gradiente se calcula la derivada parcial respecto a cada parámetro en el punto de evaluación. La derivada indica el valor y sentido en que se encuentra el mínimo más próximo. Este puede ser tanto un mínimo local como global, el método no lo puede diferenciar.</p>

<p style="text-align:justify">El resultado de la derivada se le resta a cada uno de los parámetros multiplicado por la velocidad de aprendizaje &alpha;. La velocidad de aprendizaje generalmente tiene un valor entre 0 y 1 e indica lo rápido que converge el algoritmo. Es importante notar que es necesario seleccionar un valor adecuado. Un valor demasiado bajo puede provocar que nunca se alcance el mínimo. Por otro lado, un valor lo demasiado alto podría saltarse el mínimo.</p>

<p style="text-align:justify">El gradiente es la generalización vectorial de la derivada, es un vector de tantas dimensiones como la función y cada dimensión contiene la derivada parcial en dicha dimensión:</p>

\begin{align}
    \nabla f &= [\frac{\partial f}{ \partial x_1}, \frac{\partial f}{\partial x_2}, ..., \frac{\partial f}{\partial x_n} ]
\end{align}

El gradiente &nabla;<i>f<sub>x</sub></i> es el vector que contiene la información de cuanto crece la función en un punto especifico <i>x</i> por cada dimensión de nuestra función de forma independiente. 

Se debe tomar una dirección de pasos a avanzar para descender, pero no conocemos cuánto debemos descender. Una estrategia sería si la función crece “mucho” descender mucho, y si crece “poco” descender poco. Es decir dar el paso en forma proporcional al gradiente. Para poder controlar la proporción del paso utilizaremos un parámetro <i>k</i> de tal forma que el paso a dar se cuantifica:

\begin{align}
    \alpha \nabla
\end{align}


&alpha; es un parámetro de tunning del método descenos de gradiente al que se le conoce como: learning rate (Rango de aprendizaje).

Se debe calcular el valor a dónde se llegue una posición en <i>x</i> donde se pueda avanzar a una nueva posición <i>x&#772;</i> que dependerá de lo pasos que el algoritmo avance, el tamaño y dirección: 

\begin{align}
    \bar{x} &= x - \alpha \nabla f(x)
\end{align}

Para cada nueva posición podemos medir la diferencia entre <i>x&#772;</i> y <i>x</i> si este valor es menor que cierto umbral o si se alcanza un máximo de iteraciones, se puede decidir que se ha alcanzado un punto mínimo.

```python
def Gradient(x, y, func):
    xgrad = 0
    ygrad = 0
    
    if func == "Rastrigin's":
        xgrad = 2*x + 20*np.pi*np.sin(2*np.pi*x)
        ygrad = 2*y + 20*np.pi*np.sin(2*np.pi*y)
    elif func == "De Jong's":
        xgrad = 2*x 
        ygrad = 2*y
    elif func == "Rosenbrock's Valley":
        xgrad = -400*x*(y-x**2) + 2*(1-y)
        ygrad = 200*(1-x**2)
    elif func == "Rotated Hyper-Ellipsoid":
        xgrad = 4*x
        ygrad = 2*y
    elif func == "Griewangk's":
        xgrad = (x/200) + np.sin(x)*np.cos(y/np.sqrt(2))
        ygrad = (y/200) + (np.cos(x)*np.sin(y/np.sqrt(2)))/np.sqrt(2)

    return xgrad,ygrad
```

```python
def Gradient_Descent(P0,func,gamma,precision,gnrtns):
    P = Pob2Number(P0)
    bs = []
    for I in P:
        next_step = I
        for i in range(gnrtns):
            current = next_step
            sx,sy = Gradient(current[0],current[1],func)
            next_step = [current[0] - gamma*sx, current[1] - gamma*sy]
            step = (current[0] - next_step[0], current[1] - next_step[1])
            if step[0] <= precision and step[1] <= precision:
                break
        bs.append(next_step)
    return bs
```
